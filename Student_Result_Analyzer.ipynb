```python
import pandas as pd

data = {
    "Name": ["Aditya", "Himanshu", "Raju", "Aman", "Riya", "Rahul", "Khushi", "Neha", "Pooja", "Saurav", "Kiran", "Ayush", "Vishal", "Satyam", "Ankit"],
    "Math": [78, 74, 92, 84, 86, 93, 85, 76, 81, 79, 68, 76, 94, 78, 34],
    "Science": [68, 56, 79, 96, 83, 76, 84, 82, 74, 73, 75, 69, 77, 88,98],
    "English": [97, 98, 99, 88, 89, 93, 95, 91, 86, 92, 74, 58, 76, 77, 97],
    "Hindi": [46, 42, 85, 72, 38, 73, 44, 77, 88, 99, 33, 56, 69, 73, 91]
}

df = pd.DataFrame(data)
df
```

```python
df["Total"] = df["Math"] + df["Science"] + df["English"] + df["Hindi"]
df["Average"] = (df["Total"] / 4).round(2)

df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 40 else "Fail")

df
```

```python
pass_count = (df["Result"] == "Pass").sum()
fail_count = (df["Result"] == "Fail").sum()

print("Pass count =", pass_count)
print("Fail count =", fail_count)
```

```python
topper = df.sort_values(by="Total", ascending=False).head(1)
topper
```

```python
def grade(avg):
 if avg >= 90:
  return "A+"
 elif avg >= 80:
  return "A"
 elif avg >= 70:
  return "B+"
 elif avg >= 50:
  return "C"
 else:
  return "F"

df["Grade"] = df["Average"].apply(grade)
df
```

```python
print("Math Average =", round(df["Math"].mean(), 2))
print("Science Average =", round(df["Science"].mean(), 2))
print("English Average =", round(df["English"].mean(), 2))
print("Hindi Average =", round(df["Hindi"].mean(), 2))
```

```python
df.to_csv("final_student_report.csv", index=False)
print("Report save as final_student_report.csv")
```
